```python
def vaapukkaleet():
    perms = list()
    for seed in range(pow(2,12)):
        b_1 = list()
        for c in "{0:012b}".format(seed):
            b_1.append(int(c))

        b_2 = [0,1,1,0,0,0,0,1,0,1,0,0]
        b_3 = [ b_1[i] * b_2[i] for i in range(12)]
        n = 0
        leet_index = list()
        
        for i in range(12):
            n += b_3[i]
            if b_3[i] & 1:
                leet_index.append(i)

        for i in range(pow(2, n)):
            lbin = "{0:04b}".format(i)
            lbin = lbin[-n:]      
            b = b_1.copy()
            for m in range(n):
                b[leet_index[m]] += int(lbin[m])
            
            perms.append(b)

    return perms

```

```python
p = vaapukkaleet()
```
